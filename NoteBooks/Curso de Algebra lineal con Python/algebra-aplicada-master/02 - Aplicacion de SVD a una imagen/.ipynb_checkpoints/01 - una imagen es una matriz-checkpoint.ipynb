```python
# Es importante aplciar SVD en una imagen porque tiene implicaciones en la compresión al memoneto de transmitirlas.
# O para como poder ocupar menos espacio.
# El computador ve las iamgnes como una matriz.


%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
# Traemos de PIL el módulo Image
from PIL import Image
```

```python
# Queremos usar el estilo clásico.
plt.style.use('classic')

# Abrimos la imagne en la ruta especificada.
imagen = Image.open(".\\imagenes\\imagen_ejemplo_frida_bredesen.jpg")

plt.imshow(imagen)
# ¿
```

```python
# Convertir a escala de grises.
imagen_gr = imagen.convert('LA')
print(imagen_gr)
```

```python
# Tranformamos la imagen en una matriz.( En el array hay una lista de nuestra imagen en escala de grises, y que tome todos los datos de la banda cero y me lo guarde como float)

imagen_mat = np.array(list(imagen_gr.getdata(band=0)), float)
print(imagen_mat)
# Nos devuelve una tira, un vector.
```

```python
#¿Esta sucediendo que la imagen está entre 236 y 193?.  Para comprenderlo mejor hay que transformarlo de vector a matriz.
# 
imagen_mat.shape = (imagen_gr.size[1], imagen_gr.size[0])
print(imagen_mat)
```

```python
# Para saber las dimensiones que tiene. Esta unidad esta dada en pixeles.3456, 3693, filas y columnas.
print(imagen_mat.shape)
```

```python
# le podemos decir que grafique a la matriz.!!! Le decimos que haga una conversión a la escala de grises.
plt.imshow(imagen_mat, cmap = 'gray')
plt.show()
```

```python
# Podemos alterar los valroes de la matriz. Para modificar los colores.
imagen_mat_2 = imagen_mat / 10
print(imagen_mat_2)
```

```python
# Imagen transformada de la matriz divida 10, estamos tranformando
# NOTA: ¿Es necesario conservar la escala o con tener la relación d elos valores nos alcanza.?

plt.imshow(imagen_mat_2)
plt.show()
```

```python
# ¿Qué pasa si a la matriz tranformada le decimos que lo lea como escala de grises?
# Queda comprobado que no es importante los valroes que tenemos si no la relación que tenemos entre ellos.
# En ML por optimización y para evitar convergencia, nos conviene que nuestros valores estén entre 0 y 1, podemos hacer esto dividiendo por 
# El máximo valo posible. Al leer la imagen, los valores max = 255 y mini = 0, es una forma de normalizarlo.
plt.imshow(imagen_mat_2, cmap="gray")
plt.show()
```

```python
# Np max me permite saber el máximo de la matriz original y tranformada
print(np.max(imagen_mat_2))
print(np.max(imagen_mat))
```

```python
# Np min me devuelve mínimo de la matriz original tranformada
print(np.min(imagen_mat_2))
print(np.min(imagen_mat))
```
