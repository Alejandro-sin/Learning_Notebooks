## Respuesta general
Consigo todo el documento HTML

```python
html = requests.get(url).text
soup = BeautifulSoup(html,'lxml')
print(soup.prettify())
```

### Consiguiendo Links de Noticias

En este chunk conseguiré todos los enlaces a los articulos de la página así como las categorías de las noticias.

![s](https://k60.kn3.net/taringa/0/B/6/F/2/3/MoraCaceres/4F4.jpg)




```python

#
links_articles = []

divs = soup.find('div', attrs={'class': 'article-title'}).find_all('a')
print(divs)

```

```python
# Consigo las secciones enmarcadas en un anchor:
sections = soup.find('div', attrs={'class':'p12-dropdown-column'}).find_all('a')
sections
```

```python
# Esto devuelve un :bs4.element.Tag
sec = sections[0]
# Y la notación de brakets me devuelve  el atributo: 'https://www.pagina12.com.ar/secciones/el-pais'
sec['href']

links_sections= [sec['href'] for sec in sections]
links_sections
    
#Convierto todos los nombres a strings    
name_sections=[str(sec.string)   for sec in sections]
name_sections

# Quiero guardar en un diccionario mis pares clave valor. Con fuerza bruta
dict_sections = {}
for name in name_sections:
    for link in links_sections:
        dict_sections[name] = link
dict_sections
```

```python
# Guardarlo usando dictionary comprehension:
dict_sections ={name_sections[i]: links_sections[i] for i in range(len(name_sections)) }
dict_sections
```

```python
# Usar la función zip.
dict_sections = dict(zip(name_sections, links_sections))
dict_sections
```
