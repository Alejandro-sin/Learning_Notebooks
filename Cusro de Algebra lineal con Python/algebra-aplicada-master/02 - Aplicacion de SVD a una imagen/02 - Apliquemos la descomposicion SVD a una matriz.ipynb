```python
# ¿Qué efecto tien ela descomposición SVD sobre la imagen?
# Si queremos disminuir el tamño dle archivo, tenemos que definir cuando la imagen es reconocible.


%matplotlib inline

import numpy as np

import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.style.use('classic')
imagen = Image.open(".\\imagenes\\imagen_ejemplo_frida_bredesen.jpg")
plt.imshow(imagen)
```

```python
# Se convierte en escal a de grises para poder trabajar con cualquier imagen.
imagen_gr = imagen.convert('LA')

# La convertimos en una matriz. Lista, con laimganene escala de grises y que de esta tome la data que figure en la banda = 0.
imagen_mat = np.array(list(imagen_gr.getdata(band=0)), float)
print(imagen_mat)

# Le damos la dirección correcta en esta matriz.
imagen_mat.shape = (imagen_gr.size[1], imagen_gr.size[0])
```

```python
# Una vez creada la matriz  la descomponemos en SVD Valores singulares.

U, D, V = np.linalg.svd(imagen_mat)
```

```python
print(imagen_mat.shape)
print(U.shape)
print(D.shape)
print(V.shape)
# Aqui la pregunta es, tenemos 3456 valores singulares, con cuantos deberíamos quedarnos?
```

```python
# D tiene los valores singulares, ordenados . El primer valro contiene mas ifnroamción, mas cvarianza de los datos.
# # Recunstruimos a imagen a aprtir de estos valores.
# aqui 
# i=1

imagen_recons = np.matrix(U[:, :1]) * np.diag(D[:1]) * np.matrix(V[:1,:])
plt.imshow(imagen_recons, cmap='gray')
# Volvimos a la zebra irreconocible, porque está expresada en sus valores matriciales singulares.
```

```python
# La compresión de imagenes por medio de SVD hace uso de que muy pocos de lso valroes singualres,
# relamente están represrntando información, la mayoría de llos a penas capturar un procentaje muy pequeño de su varianza

#  Reto hacer un ciclo for a ver como lo procesa.

i = 5

    imagen_recons = np.matrix(U[:, :i]) * np.diag(D[:i]) * np.matrix(V[:i,:])
    plt.imshow(imagen_recons, cmap='gray')

titulo = "valores singulares = %s" % i

plt.title(titulo)
plt.show()
```
